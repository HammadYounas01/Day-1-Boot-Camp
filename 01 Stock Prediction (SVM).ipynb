```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Trading-features.csv')
```

```python
df
```

```python
df.dtypes
```

```python
df['created_at']=pd.to_datetime(df['created_at'])
```

```python
df['created_at']
```

```python
df['time_column']=None
```

```python
df['hours']=None
```

```python
df['minutes']=None
```

```python
df['seconds']=None
```

```python
df['time_column']=df['created_at'].dt.time
```

```python
df['hours']=df['time_column'].apply(lambda x: x.hour)
```

```python
df['minutes']=df['time_column'].apply(lambda x: x.minute)
```

```python
df['seconds']=df['time_column'].apply(lambda x: x.second)
```

```python
df
```

```python
df.fillna(0,inplace=True)
```

```python
df1=df.isnull().sum()
```

```python
df1[df1>0]
```

```python
df['Prediction']=None
```

```python
df['Prediction']=pd.Series([1 if current>previous else 0 if current<previous else 2 for current, previous in zip(df['last_price'], df['last_price'].shift())])
```

```python
df['Prediction'].value_counts()
```

```python
df1=df.drop('id',axis=1)
```

```python
df1=df1.drop('created_at',axis=1)
```

```python
df1=df1.drop('Prediction',axis=1)
```

```python
df1
```

```python
df1=df1.drop('time_column',axis=1)
```

```python
df3=df['Prediction']
```

```python
X=np.asarray(df1)
```

```python
y=np.asarray(df3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn import svm
```

```python
classifier=svm.SVC()
classifier.fit(X_train,y_train)
y_predict=classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```
